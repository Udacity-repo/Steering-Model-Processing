```python
import pandas as pd
import os
import matplotlib.pyplot as plt
# Config the matlotlib backend as plotting inline in IPython
%matplotlib inline

def get_filename(folder, start):
    file = [f for f in os.listdir(folder) if f.startswith(start)][-1]
    print(file)
    return file

def plot_eval(folder, start, subplots=True, limit=500):
    file = get_filename(folder, start)
    data = pd.read_csv(os.path.join(folder, file))
    data.iloc[:limit].plot(subplots=subplots, figsize=(10, 4))
```

```python
plot_eval('checkpoints/eval', 'eval-', subplots=False)
```

```python
plot_eval('checkpoints/train', 'eval-', subplots=False)
```
